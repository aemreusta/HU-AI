```python
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Define fuzzy variables
heart_rate_ctrl = ctrl.Antecedent(np.linspace(0, 1, 100), "heart_rate")
resp_rate_ctrl = ctrl.Antecedent(np.linspace(0, 1, 100), "resp_rate")
temp_ctrl = ctrl.Antecedent(np.linspace(0, 1, 100), "temp")
wbc_ctrl = ctrl.Antecedent(np.linspace(0, 1, 100), "wbc")
lactate_ctrl = ctrl.Antecedent(np.linspace(0, 1, 100), "lactate")
sepsis_ctrl = ctrl.Consequent(np.linspace(0, 1, 100), "sepsis")

# Membership functions
for var in [heart_rate_ctrl, resp_rate_ctrl, temp_ctrl, wbc_ctrl, lactate_ctrl]:
    var["low"] = fuzz.trimf(var.universe, [0, 0, 0.5])
    var["medium"] = fuzz.trimf(var.universe, [0, 0.5, 1])
    var["high"] = fuzz.trimf(var.universe, [0.5, 1, 1])

sepsis_ctrl["no"] = fuzz.trimf(sepsis_ctrl.universe, [0, 0, 0.5])
sepsis_ctrl["yes"] = fuzz.trimf(sepsis_ctrl.universe, [0.5, 1, 1])

# Define rules
rule1 = ctrl.Rule(
    heart_rate_ctrl["high"]
    & resp_rate_ctrl["high"]
    & temp_ctrl["high"]
    & wbc_ctrl["high"]
    & lactate_ctrl["high"],
    sepsis_ctrl["yes"],
)
rule2 = ctrl.Rule(
    heart_rate_ctrl["low"]
    & resp_rate_ctrl["low"]
    & temp_ctrl["low"]
    & wbc_ctrl["low"]
    & lactate_ctrl["low"],
    sepsis_ctrl["no"],
)
# Add more rules as per your design

# Create control system
sepsis_ctrl_system = ctrl.ControlSystem([rule1, rule2])  # Add all rules
sepsis_ctrl_sim = ctrl.ControlSystemSimulation(sepsis_ctrl_system)
```

```python
def predict_sepsis(sample):
    sepsis_ctrl_sim.input["heart_rate"] = sample[input_features.index("heart_rate")]
    sepsis_ctrl_sim.input["resp_rate"] = sample[input_features.index("resp_rate")]
    sepsis_ctrl_sim.input["temp"] = sample[input_features.index("temp")]
    sepsis_ctrl_sim.input["wbc"] = sample[input_features.index("wbc")]
    sepsis_ctrl_sim.input["lactate"] = sample[input_features.index("lactate")]

    # Compute the result
    sepsis_ctrl_sim.compute()
    return sepsis_ctrl_sim.output["sepsis"]
```

```python
predictions = []
for i in range(len(X_scaled)):
    sample = X_scaled[i]
    pred = predict_sepsis(sample)
    predictions.append(pred)

# Convert predictions to binary based on a threshold (e.g., 0.5)
y_pred = [1 if p > 0.5 else 0 for p in predictions]
```

```python
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve

# Confusion Matrix
cm = confusion_matrix(y, y_pred)
tn, fp, fn, tp = cm.ravel()

# Metrics
TPR = tp / (tp + fn)  # Sensitivity, Recall
FPR = fp / (fp + tn)
TNR = tn / (tn + fp)  # Specificity
FNR = fn / (fn + tp)
F1 = 2 * tp / (2 * tp + fp + fn)
ROC_AUC = roc_auc_score(y, predictions)

print("Confusion Matrix:")
print(cm)
print(f"TPR (Recall): {TPR:.2f}")
print(f"FPR: {FPR:.2f}")
print(f"TNR (Specificity): {TNR:.2f}")
print(f"FNR: {FNR:.2f}")
print(f"F1 Score: {F1:.2f}")
print(f"ROC AUC: {ROC_AUC:.2f}")
```

```python
fpr, tpr, thresholds = roc_curve(y, predictions)
plt.figure()
plt.plot(fpr, tpr, label="FIS (area = %0.2f)" % ROC_AUC)
plt.plot([0, 1], [0, 1], "k--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.legend(loc="lower right")
plt.show()
```
