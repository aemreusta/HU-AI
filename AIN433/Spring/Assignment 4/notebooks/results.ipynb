```python
import re
import pandas as pd

# Initialize result dictionary
results = {
    "YOLOv3Tiny_8_1e-4": [],
    "YOLOv3Tiny_16_1e-4": [],
    "EfficientNet_8_1e-4": [],
    "EfficientNet_16_1e-4": [],
}

# Read log file as log_content
with open(
    "/Users/emre/GitHub/HU-AI/AIN433/Spring/Assignment 4/logs/log_train_20240523-050449.log",
    "r",
) as file:
    log_content = file.readlines()

# Define regex pattern for extracting information
pattern = re.compile(
    r"Epoch \[(\d+)/200\], Train Loss: ([\d.]+), Val Loss: ([\d.]+), Train CLS Loss: ([\d.]+), "
    r"Val CLS Loss: ([\d.]+), Train REG Loss: ([\d.]+), Val REG Loss: ([\d.]+), Train Accuracy: ([\d.]+), "
    r"Val Accuracy: ([\d.]+), Train mIoU: ([\d.]+), Val mIoU: ([\d.]+), Time: ([\d.]+)s"
)

# Extract results from log file
for line in log_content:
    match = pattern.search(line)
    if match:
        epoch = int(match.group(1))
        train_loss = float(match.group(2))
        val_loss = float(match.group(3))
        train_cls_loss = float(match.group(4))
        val_cls_loss = float(match.group(5))
        train_reg_loss = float(match.group(6))
        val_reg_loss = float(match.group(7))
        train_accuracy = float(match.group(8))
        val_accuracy = float(match.group(9))
        train_miou = float(match.group(10))
        val_miou = float(match.group(11))
        time = float(match.group(12))

        if "YOLOv3Tiny" in line:
            if "batch size: 8" in line:
                results["YOLOv3Tiny_8_1e-4"].append(
                    (
                        epoch,
                        train_loss,
                        val_loss,
                        train_accuracy,
                        val_accuracy,
                        train_miou,
                        val_miou,
                    )
                )
            elif "batch size: 16" in line:
                results["YOLOv3Tiny_16_1e-4"].append(
                    (
                        epoch,
                        train_loss,
                        val_loss,
                        train_accuracy,
                        val_accuracy,
                        train_miou,
                        val_miou,
                    )
                )
        elif "EfficientNet" in line:
            if "batch size: 8" in line:
                results["EfficientNet_8_1e-4"].append(
                    (
                        epoch,
                        train_loss,
                        val_loss,
                        train_accuracy,
                        val_accuracy,
                        train_miou,
                        val_miou,
                    )
                )
            elif "batch size: 16" in line:
                results["EfficientNet_16_1e-4"].append(
                    (
                        epoch,
                        train_loss,
                        val_loss,
                        train_accuracy,
                        val_accuracy,
                        train_miou,
                        val_miou,
                    )
                )

# Convert results to DataFrame for better visualization
dfs = {}
for key, value in results.items():
    dfs[key] = pd.DataFrame(
        value,
        columns=[
            "Epoch",
            "Train Loss",
            "Val Loss",
            "Train Accuracy",
            "Val Accuracy",
            "Train mIoU",
            "Val mIoU",
        ],
    )

# Display DataFrames
for key, df in dfs.items():
    print(f"Results for {key}:")
    display(df.head(10))
```
